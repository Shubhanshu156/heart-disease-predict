# draw multiple plot 

```python
%matplotlib inline
fig,((axis1,axis2),(axis3,axis4))=plt.subplots(nrows=2,ncols=2)
axis1.hist(a)
axis2.plot(a**2,a)
axis3.scatter(a**3,a**4)
axis4.bar(a**4,a)
```

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Date"]=pd.date_range('1/1/2022',periods=10)
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales['Price'].sum()
```

```python
car_sales["Total Sales"]=car_sales["Price"].cumsum()
```

```python
car_sales
```

```python
car_sales.plot(x='Date',y='Total Sales')
```

```python
heart_dis=pd.read_csv("heart-disease.csv")
```

```python
heart_dis.plot.hist(subplots=True,figsize=(100,100));
```

```python
heart_dis
```

```python
heart_dis.plot(x='chol',y='target',c='target',kind='scatter');
```

```python
over50=heart_dis[heart_dis['age']>50]
len(over50)
```

```python
# pyplot method
over50.plot(x='age',y='chol',c='target',kind='scatter');
```

```python
# OO Method
fix,ax=plt.subplots(figsize=(10,5))
scatter=ax.scatter(over50['age'],over50['chol'],c=over50['target'],cmap='RdBu');
ax.legend(*scatter.legend_elements(),title='Target');
ax.set_title('Heart Disease Chol vs Age');
ax.set_xlabel("Age")
ax.set_ylabel("Chol")
ax.set_xlim([50,80] )
ax.set_ylim([100,400])
ax.axhline(y=over50['chol'].mean(),color='r',linestyle='--');
```

```python
over50
```

```python
# Subplot of thalh,chol vs age
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,6),sharex=True)
scatter1=ax1.scatter(over50['age'],over50['chol'],c=over50['target'],cmap='RdBu');

ax1.set_ylabel("Chol")
ax1.legend(*scatter1.legend_elements(),title='Target');
scatter2=ax2.scatter(over50['age'],over50['thalach'],c=over50['target'],cmap='RdBu');

ax2.set_ylabel("thalach")
ax2.legend(*scatter1.legend_elements(),title='Target');
ax1.set_ylim([0,400])
ax2.set_ylim([80,200])
fig.suptitle("heart Disease Analysic",fontsize=26,fontweight=10)
```
